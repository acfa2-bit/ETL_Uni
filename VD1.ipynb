```python
import numpy as np
```

```python
# criando uma lista Python

minhalista = [0, 1, 2, 3, 4, 5, 6, 7, 8]
minhalista
```

```python
#Transformando essa lista em uma Array NumPy
#O array é multidimensional

meuarray = np.array(minhalista)
meuarray
```

```python
# Formas simplificadas de criar arrays
# Uma forma mais simples de criarmos uma lista:
minhalista = list (range(9))
minhalista

#Transformando agora essa lista em um Array Numpy
array1 = np.array(range(9))
array1

#Arange: Fazendo as etaás anteriores numa unica sintaxe:
array2 = np.arange(9)
array2
```

```python
#Start, stop, step:
array3 = np.arange(0, 10, 0.5)
array3

#Arredondando os valores obtidos
novoarray = np.around(array3)
novoarray

#Também podemos "forçar" o tipo dos elementos:
array3 = np.arange(0, 10, 0.5, dtype=int)
array3
```

```python
#Também podemos pré-determinar o tipo dos elementos
array4 = np.arange(12)

#Ainda teremos os números inteiros, mas posso deixar meu array previamente preparado para trabalhar futuramento com os tipos float (.)
array4 = np.arange(12.)
array4
```

```python
#Verificando o tipo do objeto (estrutura de dados):
type(array4)

#Verificando o tipo dos elementos internos do array
array4.dtype
```

```python
#Criação e seleção de elementos

#Também podemos criar um array digitando os valores
vetor1 = np.array([0, 1, 2, 3, 4])
print(vetor1)

#Seleção e fatiamento
#Selecionando um elemento específico
vetor1[3]

#Selecionando um intervalo específico
vetor1[0:2] #ou também print(vetor1[0:2])

#Selecionando um intervalo específico com saltos
#vetor1[1::2]
```

```python
#Criando um array de strings:
s = np.array(['Python', 'Profissional', 'Java', 'Água'])
s

#Misturando tipos (strings e números):
s = np.array(['Python', 1464322315, 'Profissional', 'Java', 'Água'])
s
```

```python
#Alteração de elementos e restrição de tipos

#Alterando um elemento array:
vetor1[2] = 10
vetor1

#Vamos tentar incluir elementos de outro tipo (string):
vetor1[0] = "novo elemento"
vetor1

#Agora tente o seguinte (float em array int):
vetor1[0] = 45.6
vetor1
```

```python
#Linspace e Matrizes
#importante para Machine Learning, Deep Learning.

arr1 = np.arange(10) # 0 a 9 (stop)
arr2 = np.arange(10, 20) # 10 a 19 (start, stop)
arr3 = np.arange(10, 20, 2) # 10 a 19 (start, stop, step)
arr4 = np.arange(10, 20, 1.4) #passo float


```

```python
# Exemplos de linspace

arr5=np.linspace(10,20,11)

arr6=np.linspace(10,20,101)

arr7=np.linspace(0, np.pi, 100)

arr5
arr6
arr7
```

```python
#Criando a primeira matriz

#Criando a primeira matriz (com quebra de linha para melhor leitura):
minhamatriz = np.array([[1, 2, 3],
                        [4, 5, 6]])
minhamatriz

#Lembrando que voce pode escrever apenas assim (na mesma linha):
minhamatriz = np.array([[1,2,3], [4, 5,6]])
minhamatriz
```

```python
# Podemos criar uma matriz a partir de uma lista de listas:
minhalista = [[13,81,22], [0, 34, 59], [21, 48, 94]]

# Agora usaremos a função np.matrix:
minhamatriz = np.matrix(minhalista)
minhamatriz
```

```python
# A função np.zeros pode formar tanto valores quanto matrizes preenchidos com zeros
# Vetor 1D
x = np.zeros(3)
x

#Matriz (2D)
x = np.zeros((3,2))
x
```

```python
# Você não precia necessariamente criar uma variável para usar a função
# mas, para armazenar e usarmos depois, temos que criar a variável, para guardarmos nela
np.zeros((4,4))
```

```python
# A função np.ones funciona da mesma forma que a np.zeros:
x = np.ones((4,4))
x

# Para preencher apenas com 8s (ou com qualquer outro número que não 0 e 1 - multiplicamos pelo do one)
x = 8*np.ones((4,4))
x

```

```python
#Matrizes diagonais e identidade

#Matrizes com valores na diagonal:
z = np.diag(np.array([1, 2, 3, 4, 5, 6]))
z

# Matriz identidade (diagonal principal com 1, resto 0):
y = np.eye(5)
y
```

```python
#Verficando o chape da matriz
#Podemos verificar o shape (formato) da matriz, que nos retorna uma tupla com o número de linhas e colunas

np.shape(minhamatriz)
```

```python
#Verificando tipos

#tipo do objeto como um todo
type(minhamatriz)

#Verificando o tipo dos dados da matriz - tipo do conteúdo
minhamatriz.dtype
```

```python
# Seleção e alteração de elementos

#Selecionando elementos da matriz:
minhamatriz[2,1] #linha indice 2, coluna indice 1

#Alterando um elemento da matriz:
minhamatriz[2,1] = 100
minhamatriz
```

```python
# Operações básicas com arrays

#soma dos elementos:
np.sum(vetor1)

#Produto dos elementos:
np.prod(vetor1)

#Soma acumulada:
np.cumsum(vetor1)
```

```python
#valores mínimo e máximo

#Valor Mínimo
#a.min()
print(np.min(vetor1))

#Valor Máximo
#a.max()
print(np.max(vetor1))
```

```python
#Estatística básica
#o Numpy pega toda a população, enquanto o pandas calcula como se fosse uma amostra (na fórmula)

#Média
np.mean(minhamatriz)

#mediana
np.median(minhamatriz)

#desvio padrão
np.std(minhamatriz)

#variância
np.var(minhamatriz)
```

```python
#Desafio final
#Crie um array com 20 elementos (você pode escolher usar arange ou linspace)
#Calcule e imprima os seguintes valores estatísticos deste array:
#média, valor mínimo, desvio padrão, valor máximo, variância

arraynova = np.arange(20)
arraynova

print(np.var(arraynova))
print(np.mean(arraynova))
print(np.std(arraynova))
print(np.min(arraynova))
print(np.max(arraynova))

```
